```python
from scipy.spatial import cKDTree
import numpy as np

def coordinate_distance_optimized():
    lst = []
    coords_1 = np.array(df1_new[['Latitude', 'Longitude']])
    coords_2 = np.array(df2_new[['Latitude', 'Longitude']])
    
    # Build KD-trees for the two sets of coordinates
    tree_1 = cKDTree(coords_1)
    tree_2 = cKDTree(coords_2)
    
    # Find pairs of coordinates that are within 0.5 km of each other
    pairs = tree_1.query_ball_tree(tree_2, r=0.5, p=2.0)
    
    # Iterate over the pairs and calculate their distances
    for i, pair in enumerate(pairs):
        for j in pair:
            dist = geopy.distance.geodesic(coords_1[i], coords_2[j]).km
            lst.append(dist)
    
    return lst
```

```python
import pulp

# Define the problem
problem = pulp.LpProblem("Bus Stop Assignment", pulp.LpMinimize)

# Define the decision variables
z = {c: pulp.LpVariable(f"z_{c}", cat="Binary") for c in items}
y = {(p, c): pulp.LpVariable(f"y_{p}_{c}", cat="Binary") for p in people for c in items}

# Define the objective function
problem += pulp.lpSum([z[c] + d.get((p, c), 0) * y[(p, c)] for p in people for c in items])


# Define the constraints
for p in people:
    problem += pulp.lpSum([y[p, c] for p in people for c in items]) == 1

for c in items:
    for p in people:
        problem += y[p, c] <= z[c]

problem += pulp.lpSum([z[c] for c in items]) == 3

# Solve the problem
status = problem.solve()

# Print the solution
if status == pulp.LpStatusOptimal:
    print("Optimal solution found")
    for c in items:
        if z[c].value() == 1:
            print(f"Item {c} is selected")
else:
    print("No optimal solution found")
```
